```python
import treecorr
import numpy as np
import matplotlib.pyplot as plt
import os
from astropy.io import fits
```

# Load Catalog

```python
datadir_root = '/scratch/network/jm8767'
datadir = os.path.join(datadir_root, 'y3kp_cats')
desi_version = 'kibo'
datadir_desi = os.path.join(datadir_root, f'desi/spectro/redux/{desi_version}/zcatalog/v1')
outdir = os.path.join(datadir_root, 'cl-lens')
outfile = os.path.join(outdir, f'des_y3_v0.50_v2026-03-12_{desi_version}.hdf5') 
infile_desi = os.path.join(datadir_desi, f'zall-tilecumulative-{desi_version}.fits')
```

```python
fits_file = infile_desi
```

```python
with fits.open(fits_file) as hdul:
    # Print a list of all column names in the ZCATALOG extension
    print(hdul[1].columns.names)
```

```python
hdul = fits.open(fits_file)
# 2. See what is inside
# This prints a summary of the extensions. 
# Typically, extension 0 is the Primary header (often empty for catalogs), 
# and extension 1 contains the actual Binary Table data.
hdul.info()

# 3. Access the data payload (assuming it's in extension 1)
data = hdul[1].data

# 4. Extract specific columns into numpy arrays
# You can use the exact column names as they appear in the FITS header
ra = data['RA']
dec = data['DEC']
g1 = data['g1']  # or whatever the shear column is named (e.g., 'e1', 'shear1')
g2 = data['g2']

# 5. Always close the file when you are done to free up memory
hdul.close()
```

# TreeCorr Calculations

https://rmjarvis.github.io/TreeCorr/_build/html/gg.html

```python
cat = treecorr.Catalog(ra=ra, dec=dec, g1=g1, g2=g2, ra_units='degrees', dec_units='degrees')
gg = treecorr.GGCorrelation(min_sep=1.0, max_sep=100.0, nbins=20, sep_units='arcmin')
gg.process(cat)

r = np.exp(gg.meanlogr)

# correlation functions
#xi_+
xip = gg.xip
# # xi_-
xim = gg.xim

#calculate variance
sig_xip = np.sqrt(gg.varxip)
sig_xim = np.sqrt(gg.varxim)
```

# Plot

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.errorbar(r, xip, yerr=sig_xip, label=r'$\xi_+$', color='navy', capsize=3)
ax.errorbar(r, xim, yerr=sig_xim, label=r'$\xi_-$', color='crimson', capsize=3)

ax.set_xscale('log')
ax.set_yscale('symlog', linthresh=1e-6) 
ax.set_xlabel(r'$\theta$ [arcmin]', fontsize=14)
ax.set_ylabel(r'Shear Correlation $\xi_{\pm}(\theta)$', fontsize=14)
ax.legend(fontsize=12)
ax.grid(True, which='both', linestyle=':', alpha=0.7)

plt.tight_layout()
plt.show()
```
